```python
import os
import torch
from transformers import AutoTokenizer
```

```python
path = os.path.expanduser("../models/Qwen/Qwen3-0.6B/")
tokenizer = AutoTokenizer.from_pretrained(path)
```

```python
prompts = [
    "Hello how are you",
    "What is the capital of Paris"
]

prompts = [
    # Transform to <|im_start|>user\n{text}<|im_end|>\n format
    tokenizer.apply_chat_template(
        [{"role": "user", "content": prompt}],
        tokenize=False,
        add_generation_prompt=True, # <- Add {"role": "assistant"", ...} at the end (it will be then transformed by apply_chat_template)
    )
    for prompt in prompts
]

print(prompts)
```

```python
# returns input_ids and attention_mask
inputs = tokenizer(
    prompts,
    return_tensors="pt", # <- torch.Tensor type, by default will return List object
    padding=True
)

print(inputs)
```
