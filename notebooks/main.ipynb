## Load Config

```python
from dataclasses import dataclass, fields
```

```python
@dataclass
class Config:
    model_name: str = "Hello"
    model_type: str = "World"
    model_dir: str = "Yay"
    
print(fields(Config))
# get Config's all possible field name via
print(set(field.name for field in fields(Config)))
```

```python
def load_params(**kwargs):
    config_field = set(field.name for field in fields(Config))
    
    # Use input to replace defaultr values of config
    config_kwargs = {k: v for k, v in kwargs.items() if k in config_field}
    print(config_kwargs)
    
    config = Config(**config_kwargs) # The model_name is replaced by user value, while the other two fields are still default values
    print(config)
    
load_params(model_name="Qwen3")
```
